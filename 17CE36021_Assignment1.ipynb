> we are using rational quadratic covariance function.

> d is the distance and α  and k are non negative parameters






```
def RQCovariance(x1,x2,α=1, k=1):
  distance=(x1-x2)**2
  distance=distance/(2.0*α *k)
  distance=distance+1
  return 1.0/distance
'''
here we are taking α as 1 and k as 1
'''
```

```
S=20*20
T=100
'''
20->rows and 20->col of grid
T-> TimeStep for Temporal Dimension
'''
```

```
Xs=np.arange(S)
Us0=np.zeros(S)  # Initialising  Spatial Us

Ks0=np.array([[RQCovariance(x1,x2) for x2 in Xs] for x1 in Xs]) #Covariance matrix
Us=np.random.multivariate_normal(Us0,Ks0) # spatial Normal Distribution with Covariance matrix
'''
```

```
X=[]
μ=Ust+ηst
print(μ.shape)
σ=np.ones(μ.shape[0])
for t in range(T):
  Xt=np.random.normal(μ[:,t] , σ[t])
  X.append(Xt)
X=np.array(X)
X=X.T
X.shape
```

```
import seaborn as sns
hm=sns.heatmap(X[:,14].reshape(20,-1),cbar=True,square=True,fmt='.6f')
```

A variogram value at a given h is the average squared difference between the values of the paired locations. If two locations, u and u + h, are close to each other in terms of the distance measurement, two values are similar, so the difference in their values, Z- Z, will be small.

```
Variogram=np.zeros((S,S))
Distance=np.zeros((S,S))
for i in range(S):
  for j in range(i,S):
    Cov_Mtx=np.cov(X[i],X[j])
    EXiXj=Cov_Mtx[0][0]+Cov_Mtx[1][1]-Cov_Mtx[0][1]-Cov_Mtx[1][0]
    Variogram[i][j]=Variogram[j][i]= 0.5*EXiXj
    Distance[i][j]=Distance[j][i]=np.power( ( int(i/20) - int(j/20) )**2 + ( int(j/20) - int(i/20) )**2 , 0.5)
```

```
hmt=sns.heatmap(Variogram,cbar=True,square=True, fmt='0.5')
```
